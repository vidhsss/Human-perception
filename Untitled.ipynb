```python
def visually_active_neurons(features,group,VAR_THREHOLD):
    nL=len(features[0])-1
    Nactive_neurons=np.zeros((nL,1))
    active_neurons=[None]*nL

    for layerid in range(nL):
            nNeurons=len(features[0][layerid].flatten())
            print(nNeurons)
            nimages=len(features)
            nshape=len(features)
            response=np.zeros((nNeurons,nimages))
            for ind in range(nimages):
                response[:,ind]=np.ravel(np.array(features_singleton[ind][layerid])) 
            rg1 = response[:, group[:, 0]]
            rg2 = response[:, group[:, 1]]
            rg3 = response[:, group[:, 2]]
            print(rg1.shape)
            print(np.var(rg1,axis=1).shape)
            van=np.where(np.var(rg1,axis=1)>VAR_THREHOLD &np.var(rg2,axis=1)>VAR_THREHOLD &np.var(rg3,axis=1)>VAR_THREHOLD); 
            Nactive_neurons[layerid]=len(van);
            active_neurons[layerid]=van;
    return active_neurons, Nactive_neurons
```

```python
def divisive_normalization(features_pairs, features_singleton, active_neurons, Nactive_neurons, imgpairs, imgsingles, Blank_ID):
    nL = len(features_singleton[0]) - 1
    activeLayers = [i for i in range(len(Nactive_neurons)) if Nactive_neurons[i] != 0]
    model_coefficient = np.zeros((2, nL))
    for layers in activeLayers:
        print(f'\n Layers {layers}')
        Fp = np.zeros((Nactive_neurons[layers], len(imgpairs)))
        Fsum = np.zeros((Nactive_neurons[layers], len(imgpairs)))

        for ind in range(len(imgpairs)): 
            f = [None]*3
            count = 0
            for p in range(3): 
                if imgpairs[ind][p] != Blank_ID:
                    count += 1
                    sel_index = (imgsingles[:, p] == imgpairs[ind][p])
                    temp = features_singleton[sel_index]
                    f[count-1] = temp[layers].flatten()
            temp = features_pairs[ind][layers].flatten()
            Fp[:, ind] = temp[active_neurons[layers]]
            temp_sum = np.zeros(f[0].shape)
            for i in range(count):
                temp_sum += f[i]
            Fsum[:, ind] = temp_sum[active_neurons[layers]]
        nc_combined = find_normalization_combined(Fp, Fsum)
        model_coefficient[:, layers] = nc_combined
    return model_coefficient
```

```python
def find_normalization_combined(Fcombined, Fsum):
    fc_max = np.max(Fcombined.flatten())
    fsum_max = np.max(Fsum.flatten())
    C = max(fc_max, fsum_max)
    coeff_combined, _, _, _ = np.linalg.lstsq(np.c_[Fsum/C, np.ones(len(Fsum))], Fcombined/C, rcond=None)
    return coeff_combined
```

```python
def sparseness(r):
    if (r < 0).any():
        raise ValueError('r has negative values!')
    r = r.ravel()
    n = r.shape[0]  # total number of stimuli
    s = (1 - (1/n)*np.sum(r, axis=1)**2 / np.sum(r**2, axis=1))/(1-1/n) # sparseness
    k = n - (n-1)*s  # effective number of stimuli
    return s, k
```

```python
import numpy as np
from scipy.spatial.distance import pdist

def check_global_local(features,imagepairDetails):
    nL=len(features[0])-1
    N=len(features)
    g1=imagepairDetails[:,0]
    l1=imagepairDetails[:,1]
    g2=imagepairDetails[:,2]
    l2=imagepairDetails[:,3]
    mean_global_distance=np.zeros(nL)
    mean_local_distance=np.zeros(nL)
    indexG = np.where(l1==l2)
    indexL = np.where(g1==g2)
    for Layer in range(nL):
        temp=np.array(features[0][Layer])
        fl=len(temp.flatten())
        layerF=np.zeros((fl,N))
        for ind_img in range(N):
            layerF[:,ind_img]=np.array(features[ind_img][Layer]).flatten()
        layerwiseDist = pdist(layerF, metric = 'euclidean')
        mean_global_distance[Layer]=np.nanmean(layerwiseDist[indexG])
        mean_local_distance[Layer]=np.nanmean(layerwiseDist[indexL])
    mi_global_local=(mean_global_distance-mean_local_distance)/(mean_global_distance+mean_local_distance)
    return mi_global_local




```
